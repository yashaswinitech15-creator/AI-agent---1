```python
from google.colab import userdata
GROQ_KEY = userdata.get('GROQ_API_KEY')
SEARCH_API_KEY = userdata.get('TAVILY_API_KEY')
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="llama-3.1-8b-instant", api_key=GROQ_KEY)
response = model.invoke("What is your knowledge cutoff date")
print(response.content)
```

```python
from langchain_community.tools import TavilySearchResults

search = TavilySearchResults(tavily_api_key=SEARCH_API_KEY)
tools_list = [search]
```

```python
from langgraph.prebuilt import create_react_agent

agent_executor = create_react_agent(model, tools_list)
```

```python
response = agent_executor.invoke({"messages": [("user", "Who is the current deputy CM of Andhra Pradesh?")]})
print(response["messages"][-1].content)
```
